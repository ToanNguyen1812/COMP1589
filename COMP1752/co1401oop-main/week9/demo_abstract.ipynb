```python
# ABC is the base class for Abstract Classes, and abstractmethod is a decorator to declare abstract methods.
from abc import ABC, abstractmethod 
```

```python
class Vehicle(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod     # declare move() as an abstract method
    def move(self):
        pass
```

```python
# abstract class Vehicle cannot be instantiated
v = Vehicle("Generic Vehicle")  # This will raise an error
```

```python
class Motorcycle(Vehicle):
    def __init__(self, name):
        super().__init__(name)

    # override the abstract method move() => move is no longer abstract in Motorcycle
    def move(self):
        return f"{self.name} is moving on 2 wheels with engine."
```

```python
motor = Motorcycle("Harley Davidson")
print(motor.move())
```

```python
class Car(Vehicle):
    def __init__(self, name):
        super().__init__(name)

    def move(self):
        return f"{self.name} is moving on 4 wheels with engine."
```

```python
car = Car("Toyota Camry")
print(car.move())
```

```python
class Bike(Vehicle):
    def __init__(self, name):
        super().__init__(name)

    def move(self):
        return f"{self.name} is moving on 2 wheels with human power."
```

```python
bike = Bike("Giant Escape 3")
print(bike.move())
```

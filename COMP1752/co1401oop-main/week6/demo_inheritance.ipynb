```python
# Syntax to inherit from a class in Python: class B(A) => class B inherits from class A
# class B will have access to all the methods and properties of class A, and can also have its own methods and properties.

# Example of a class Person with a method display_info, 2 attributes name and age, and a constructor to initialize these attributes.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)  # Call the constructor of the parent class (Person)
        self.grade = grade

    # Override the display_info method to include the grade information
    def display_info(self):
        super().display_info()  # Call the method from the parent class
        print(f"Student Grade: {self.grade}")

    def rank(self):
        if self.grade >= 80:
            return "Disstinction"
        elif self.grade > 60:
            return "Merit"
        elif self.grade >= 40:
            return "Pass"
        else:
            return "Fail"
```

```python
john = Student("John Doe", 20, 85)
john.display_info()  # This will call the display_info method from the Person class
print(f"John's Rank: {john.rank()}")  # This will call the rank method from the Student class
```

```python
class Animal:
    def __init__(self, kind, age):
        self.kind = kind
        self.age = age

    def display(self):
        print(f"Kind: {self.kind}, Age: {self.age}")
```

```python
class Pet(Animal):
    def __init__(self, kind, age, owner):
        super().__init__(kind, age)  # Call the constructor of the parent class (Animal)
        self.owner = owner

    def display(self):
        super().display()  # Call the method from the parent class
        print(f"Pet Owner: {self.owner}")

    def savage(self):
        self.owner = "No owner"
        print(f"The pet is now savage and has no owner.")
```

```python
dog = Animal("Dog", 5)
dog.display()  # This will call the display method from the Animal class
```

```python
dog_pet = Pet("Dog", 5, "Alice")
dog_pet.display()  # This will call the display method from the Pet class, which also
```

```python
dog_pet.savage()
```

```python
dog_pet.display()  # This will show that the pet is now savage and has no owner
```

```python
class Puppy(Pet):
    def __init__(self, age, owner):
        super().__init__("Dog", age, owner)
    
    def greet(self):
        print(f"The puppy greets its owner {self.owner} with a wagging tail!")
```

```python
class Kitty(Pet):
    def __init__(self, age, owner):
        super().__init__("Cat", age, owner)
    
    def greet(self):
        print(f"The kitty greets its slave {self.owner} with ignoring behavior!")
```

```python
ki = Kitty(3, "Bob")
ki.display()
ki.greet()
```

```python
poo = Puppy(1, "Charlie")
poo.display()
poo.greet()
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name # Private attribute
        self.age = age   # Public attribute

    def display_info(self):
        print(f"Name: {self.__name}, Age: {self.age}")
```

```python
class Student(Person):
    def __init__(self, name, age):
        super().__init__(name, age)

    def display_info(self):
        print(f'Age: {self.age}')
        print(f'Name: {self.__name}')
```

```python
john = Student("John Doe", 20)
john.display_info()
```

```python
john.age = 21
john.display_info()
```

```python
john.__name = "Paul"
john.display_info()
```

```python
try:
    a = int(input("Enter a: "))
    b = int(input("Enter b: "))
    c = a / b
    print(f"{a} / {b} = {c}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
except ValueError:
    print("Error: Invalid input! Please enter an integer.")

print('Ending...')
```

```python
try:
    a = [2, 9, 1, 10, 8, 3]
    i = int(input("Enter index: "))
    print(f"Index: {i}")
    print(f"a[{i}] = {a[i]}")
except IndexError:
    print("Error: Index out of range!")
except ValueError:
    print("Error: Invalid input! Please enter an integer.")

print('Ending...')
```

```python
# use default value
try:
    a  = { 'hello': 'xin chào', 'bye': 'tạm biệt', 'cat': 'con mèo', 'dog': 'con chó', 'unknown': 'từ không xác định' }
    en_word = input('Enter an English word: ')
    print(f"English word: {en_word}")
    print(f"Vietnamese word: {a[en_word]}")
except KeyError:
    # key not found, use default value
    print(f"Vietnamese word: {a['unknown']}")

print('Ending...')
```

```python
invalid = True
while invalid:
    try:
        a = int(input("Enter a: "))
        b = int(input("Enter b: "))
        c = a / b
        invalid = False
    except ZeroDivisionError:
        print("Error: Cannot divide by zero! Please try again.")
    except ValueError:
        print("Error: Invalid input! Please enter an integer.")
    
    if invalid:
        print("Please try again.")

print(f"{a} / {b} = {c}")
```
